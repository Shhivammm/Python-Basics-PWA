1. What is Python, and why is it popular?
  -  Python is a programming language that’s really easy to learn and use. It’s
     popular because of its simple syntax, versatility, and huge community support. People use it for web development, automation, data science, artificial intelligence, and much more.


2. What is an interpreter in Python?
  - An interpreter is a program that reads and executes Python code line by
  line. Unlike compiled languages, Python doesn’t need to be converted into machine code before running—it just executes directly.


3. What are pre-defined keywords in Python?
  - These are special words in Python that have a fixed meaning and can’t be used as variable names. Examples include if, else, while, for, def, return, and import.


4. Can keywords be used as variable names?
  - No, Python won’t allow you to use keywords as variable names because they already have a specific function in the language.


5. What is mutability in Python?
  - Mutability refers to whether an object can be changed after it’s created. If it’s mutable, like a list, you can modify its contents. If it’s immutable, like a string or a tuple, you can’t change it after it’s created.


6. Why are lists mutable, but tuples are immutable?
  - Lists are designed to be flexible, so you can add, remove, or change elements. Tuples, on the other hand, are meant to be constant, which makes them faster and safer when you don’t want data to change accidentally.


7. What is the difference between == and is operators in Python?
  - == checks if two things have the same value, while is checks if they are the exact same object in memory. For example:
  a = [1, 2, 3]  
  b = [1, 2, 3]  
  print(a == b)  True, because they have the same values  
  print(a is b)  False, because they are different objects in memory




8. What are logical operators in Python?
  - Logical operators help you make decisions in your code. They are: and (both conditions must be True) or (at least one condition must be True) not (reverses the truth value)



9. What is type casting in Python?
  - Type casting is when you change a variable from one type to another, like converting a string to an integer. For example: num_str = "10"  num_int = int(num_str) # Now it's an integer


10. What is the difference between implicit and explicit type casting?
  - Implicit type casting happens automatically, like when you add an integer to a float, and Python converts the integer to a float.
Explicit type casting is when you manually change the type, like using int(), float(), or str().


11. What is the purpose of conditional statements in Python?
  - Conditional statements (if, elif, else) let your program make decisions based on conditions. They’re super useful for controlling the flow of your code.


12. How does the elif statement work?
  - elif (short for "else if") lets you check multiple conditions. If the first if condition isn’t met, Python moves to the elif condition and checks that. If none of them are true, it moves to else. Example:



age = 18  
if age < 18:  
    print("You're a minor.")  
elif age == 18:  
    print("You're exactly 18!")  
else:  
    print("You're an adult.")

13. What is the difference between for and while loops?
  - for loops are used when you know how many times you want to repeat something. They’re great for looping over lists or ranges.

  - while loops are used when you don’t know exactly how many times you’ll need to loop, but you have a condition that needs to be met.


14. Describe a scenario where a while loop is more suitable than a for loop.
  - A while loop is better when you don’t know in advance how many times something will repeat. For example, if you’re asking a user for input and want to keep asking until they enter something valid:



user_input = ""  
while user_input.lower() != "yes":  
    user_input = input("Do you want to continue? (yes/no): ")  
print("Great! Moving forward.")
